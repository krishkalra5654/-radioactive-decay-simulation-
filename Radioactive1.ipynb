```python
import numpy as np
import matplotlib.pyplot as plt

#parameters
initial_amount = 1000
decay_constant = 0.1
time = np.linspace(0,50,500)

#Radioactive decay model
def decay_model(t , NO , lambda_):
  return NO *np.exp(-lambda_ * t)

#generate synthetic data
decayed_amount = decay_model(time , initial_amount , decay_constant)

#plot
plt.figure(figsize=(10 , 6))
plt.plot(time , decayed_amount , label="Radioactive Decay")
plt.xlabel('Time')
plt.ylabel('Number of atoms')
plt.title('Radioactive Decay of atoms')
plt.legend()
plt.grid(True)
plt.show()
```
